```
# 8.
sns.distplot(data['matchDuration']);
```

The match duration distribution is centered around 1375 and 1875(the two peaks), indicating that most matches played last between 1375 and 1875 mins/seconds (assuming the unit for match duration as it is not explicitly mentioned). It is **Bimodal**. This distribution also shows that since most of the distribution is centered around the two modes 1375 and 1875 the average match duration will also lie between 1375 and 1875 and will be closer to 1375. We can see that this is true as the mean value obtained for the match duration is 1575; which lies between 1375 and 1875 and is closer to 1375.
